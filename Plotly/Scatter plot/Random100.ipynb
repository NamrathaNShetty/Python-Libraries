* @Title: program to draw line and scatter plots for random 100 x and y coordinates

```python
import numpy as np
import pandas as pd
import plotly.express as px
from pylab import randn

x = randn(100)
y = randn(100)
df = pd.DataFrame({
    'col_x':x, 'col_y':y
})
#df.head()

fig = px.scatter(df, x='col_x', y='col_y',width=700, height=500,title="2D Scatter Plot")
fig.show()
```

```python
fig = px.line(df, x='col_x', y='col_y',width=700, height=500,title="2D Scatter Plot")
fig.show()
```
