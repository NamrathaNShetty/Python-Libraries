* @Title: Python programming to display a horizontal bar chart of the popularity of programming Languages.

```python
import matplotlib.pyplot as plt
import numpy as np

languages = ('Java','Python', 'PHP', 'Javascript', 'C#', 'C++')
y = np.arange(len(languages))
print(y)
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.barh(y,popularity,color='magenta')
plt.yticks(y, languages)

plt.xlabel('Popularity')
plt.ylabel('Languages')

plt.title('Most loved programming languages')
plt.show()
```
